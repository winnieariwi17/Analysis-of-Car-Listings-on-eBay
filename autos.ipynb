DEALING WITH INCORRECT REGISTRATION YEAR DATA

A car can't be registered after the first listing was made, therefore any car registered after the year 2016 should be filtered out. For this analysis, we will use our earliest year as 1900

```python
(~autos["registration_year"].between(1900,2016)).sum()/autos.shape[0]
```

Vehicles that fall outside of the range of 1900 and 2016 are less than 4% thereofre we can drop those columns

```python
autos= autos[autos["registration_year"].between(1900,2016)]
autos["registration_year"].value_counts(normalize=True).head(10)
```

Most vehicles are registered between the year 1998 and 2007

EXPLORING PRICE BY BRAND

```python
autos["brand"].value_counts(normalize= True)
```

Volkswagen, BMW,Opel, Mercedes,Audi and Ford each individually account for 5% of the brands. Therefore we will limit our analysis to these listings.

```python
brand_counts=autos["brand"].value_counts(normalize= True)
top_brands= brand_counts[brand_counts>.05].index
print(top_brands)
```

```python
brand_mean_prices={}

for brand in top_brands:
    top_brand_only=autos[autos["brand"]==brand]
    mean_price= top_brand_only["price"].mean()
    brand_mean_prices[brand]= int(mean_price)

brand_mean_prices
```

While volkswagen accoints for a large percentage of sales, Audi has the highest price with a mean price of 9166. Opel offers the cheapest mean price of 3028 while Volkswagen has a mid price range of 5452. This explain why it has the highest sales as it is not too cheap or too expensive.

STORING AGGREGATE DATA IN A DATA FRAME

```python
bmp_series= pd.Series(brand_mean_prices)
pd.DataFrame(bmp_series, columns=["mean_price"])
```

```python
brand_mean_mileage= {}

for brand in top_brands:
    top_brands_only=autos[autos["brand"]==brand]
    mean_mileage= top_brands_only["odometer_km"].mean()
    brand_mean_mileage[brand]= int(mean_mileage)

mean_prices= pd.Series(brand_mean_prices).sort_values(ascending= False)
mean_mileage=pd.Series(brand_mean_mileage).sort_values(ascending=False)

brand_mean_mileage
```

```python
brand_info=pd.DataFrame(mean_mileage,columns=["mean_mileage"])
brand_info
```

```python
brand_info["mean_price"]= mean_prices
brand_info
```

The car mileage does not vary as much in comparison to the car prices. However, more expensive cars tend to have a higher mileage
